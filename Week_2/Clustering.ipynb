```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## Train Model: **KMeans Clustering**

```python
# Generate synthetic clustering dataset
X, y = make_blobs(n_samples=500, centers=3, random_state=42)

# Train KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Predictions (cluster labels)
labels = kmeans.predict(X)

print("First 5 Cluster Labels:", labels[:5])
print("Cluster Centers:\n", kmeans.cluster_centers_)
```

## Plot Graph: **Scatter Plot**

```python
# Plot clusters + centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30, label="Data Points")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c="red", marker="X", s=200, label="Centers")
plt.legend()
plt.show()
```
